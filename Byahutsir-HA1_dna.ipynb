#### Create a DNA string of 10,000 characters and count ATGC and all their combinations in two and three characters.
This routine will import the following routines - random and collections. It will use some routines: random (to get random number or string) counter ( to count number of repititions of a given number or string or their combinations)


```python
import random as rn
R=[rn.randint (0, 100) for i in range (200)]
print(R)

```

```python
import collections
c= collections.Counter(R)
print (c)
```

```python
R=[rn.choice(['A','T','G','C']) for i in range (10000)]
print(R)
```

```python
DNA = ''.join(R)
print(DNA)
```
